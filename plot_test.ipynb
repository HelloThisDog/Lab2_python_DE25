- Following an example given to me by kokchun to help me with drawing shapes on a graph

```python
from circle import Circle
from matplotlib import patches
```

```python
c1 = Circle(2, 3, 1)

fig, ax = plt.subplots(1)

circle_patch1 = patches.Circle((c1.x, c1.y), radius = c1.radius)

ax.add_patch(circle_patch1)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid()
```
