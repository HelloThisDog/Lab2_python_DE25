```python
from rektangle import Rektangle
from circle import Circle
import matplotlib.pyplot as plt
from matplotlib import patches
```

## Rektangle first

- the width and length got reversed in the class, so thts why its slightly weird
- not gonna lie, its fun putting in the patches on the graph

```python
r1 = Rektangle(2, -6, 3, 7)
r2 = Rektangle(-5, 4, 6, 3)
r3 = Rektangle(-10, -5, 1, 2)
r4 = Rektangle(0, 0, 3, 4)

fig, ax = plt.subplots()

rekt_patch1 = patches.Rectangle((r1.x, r1.y), width = r1.height, height = r1.length)
rekt_patch2 = patches.Rectangle((r2.x, r2.y), width = r2.height, height = r2.length, fill = False)
rekt_patch3 = patches.Rectangle((r3.x, r3.y), width = r3.height, height = r3.length, facecolor = "red")
rekt_patch4 = patches.Rectangle((r4.x, r4.y), width = r4.height, height = r4.length, facecolor = "green")

ax.add_patch(rekt_patch1)
ax.add_patch(rekt_patch2)
ax.add_patch(rekt_patch3)
ax.add_patch(rekt_patch4)

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.grid()

plt.title("rektangle graph")
```

## Circles

```python
c1 = Circle(0, 0, 3)
c2 = Circle(-3, 4, 2)
c3 = Circle(6, 0, 7)
c4 = Circle(-10, 9, 1)

fig, ax = plt.subplots()

circle_patch1 = patches.Circle((c1.x, c1.y), radius = c1.radius)
circle_patch2 = patches.Circle((c2.x, c2.y), radius = c2.radius, fill = False)
circle_patch3 = patches.Circle((c3.x, c3.y), radius = c3.radius, edgecolor = "purple", fill = False)
circle_patch4 = patches.Circle((c4.x, c4.y), radius = c4.radius, facecolor = "pink")

ax.add_patch(circle_patch1)
ax.add_patch(circle_patch2)
ax.add_patch(circle_patch3)
ax.add_patch(circle_patch4)

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.grid()

plt.title("Circle graph")
```
